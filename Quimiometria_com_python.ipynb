###**OneWay ANOVA**


O teste One Way ANOVA é aplicado à grupos de dados. Por isso, cada grupo precisa ser passado individualmente ára que a função gere o resultado do teste.

https://www.statology.org/one-way-anova-python/

* ANOVA para medidas repetidas

https://www.statology.org/repeated-measures-anova-python/


A one-way ANOVA uses the following null and alternative hypotheses:

H0 (null hypothesis): μ1 = μ2 = μ3 = … = μk (all the population means are equal)

H1 (null hypothesis): at least one population mean is different from the rest

Since the p-value is not less than 0.05, we fail to reject the null hypothesis. This means we do not have sufficient evidence to say that there is a difference in exam scores among the groups

O dados representam análises de pH, fenóis totais, acidez e grau brix realizadas em triplicata a partir de 10 amostras de caju.

[**Pré-processamento**](https://scikit-learn.org/stable/modules/preprocessing.html#)

A padronização de conjuntos de dados é um requisito comum para muitos estimadores de aprendizado de máquina implementados no scikit-learn; eles podem se comportar mal se os recursos individuais não se parecem mais ou menos com dados normalmente distribuídos padrão: Gaussiano com média zero e variância unitária.

Na prática, muitas vezes ignoramos a forma da distribuição e apenas transformamos os dados para centralizá-los removendo o valor médio de cada recurso e, em seguida, dimensionamos dividindo os recursos não constantes por seu desvio padrão.

Por exemplo, muitos elementos usados ​​na função objetivo de um algoritmo de aprendizagem (como o kernel RBF de Support Vector Machines ou os regularizadores l1 e l2 de modelos lineares) assumem que todos os recursos estão centrados em torno de zero e têm variância na mesma ordem. Se uma característica tem uma variância que é ordens de magnitude maior do que outras, ela pode dominar a função objetivo e tornar o estimador incapaz de aprender com outras características corretamente como esperado.

O módulo de pré-processamento fornece a classe de utilitário [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler), que é uma maneira rápida e fácil de realizar a seguinte operação em um conjunto de dados do tipo array:





```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
#rodando o modelo de PCA sem pré-processamento dos dados
X = dados_FQ.loc[:, ['pH', 'Fenóis Totais', 'Acidez', 'Grau Brix']]
pca = PCA(n_components=2)
pca.fit(X)
PCA(n_components=2)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
#pré-processamento
from sklearn.preprocessing import StandardScaler

#criando uma lista com os nomes das variáveis
features = ['pH', 'Fenóis Totais', 'Acidez', 'Grau Brix']

#separando as variáveis
x = dados_FQ.loc[:, features].values

#separando os rótulos das amostras
y = dados_FQ.loc[:, ['Amostras de Caju']].values

#padronizando as variáveis
x=StandardScaler().fit_transform(x)
```

```python
#gerando a PCA com o numero de componentes definida

pca = PCA(n_components=4) #definindo número de componentes

principalComponents = pca.fit_transform(x) #transformando os dados

principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2', 'PC3', 'PC4']) #criando um novo df com dimensão reduzida
```

```python
#concatenando os dados com os rótulos
finalDf = pd.concat([principalDf, dados_FQ[['Amostras de Caju']]], axis=1) 
finalDf.head()
```

```python
#eliminando a informação da replicata dos rotolos das amostras
finalDf['Amostras de Caju'] = finalDf['Amostras de Caju'].str[:3].str.replace(".", "")
```

```python
#obtendo a variância explicada
PC1, PC2, PC3, PC4 = pca.explained_variance_ratio_
print(f'A variância explicada de cada componente principal é: \nPC1: {PC1*100}%\nPC2: {PC2*100}%\nPC3: {PC3*100}%\nPC4: {PC4*100}%')
```

```python
#plotando gráfico de scores
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC1 (60.97%)', fontsize = 15)
ax.set_ylabel('PC2 (23.77%)', fontsize = 15)
ax.set_title('Gráfico de Scores de PCA', fontsize = 20)

targets = ['A6', 'B9', 'C76', 'C10', 'B89', 'B65', 'B26', 'B74', 'B75', 'E50']
colors = ['lightpink', 'lightsteelblue', 'lightslategrey', 
          'lightblue', 'lightcyan', 'lightseagreen', 
          'lightgreen', 'darkgoldenrod', 'darkorange', 
          'darkred']

for Amostras, color in zip(targets, colors):
    indicesToKeep = finalDf['Amostras de Caju'] == Amostras
    ax.scatter(finalDf.loc[indicesToKeep, 'PC1'], finalDf.loc[indicesToKeep, 'PC2'], c = color, s=50)
ax.legend(targets)
ax.grid()
```

```python
#visualizando a tabelas de componentes e variáveis que formam os loadings
componentes = pca.components_
componentes
```

```python
#criando um novo dataframe de loadings
load = pd.DataFrame(data=componentes, columns=features, index=['PC1', 'PC2', 'PC3', 'PC4'])
load
```

Um array é um arranjo de dados que representa linhas e colunas.
Cada lista (representada pelos colchetes) representa uma linha e cada linha corresponde a uma componente principal. As colunas represetam cada uma das variáveis, como representado abaixo.

`Array(componentes, variáveis)`



|  | **variavel1** | **variavel2** | **variavel3** | ... | **variaveln** |
|:------------:|:----------:|:-----------:|:-----------:|-----|:-------:|
|       **PC1**      |        |        |        | ... |       |
|       **PC2**      |        |        |        | ... |       |
|       **PC3**      |        |        |        | ... |       |
|       **...**      |        |        |        | ... |       |
|       **PCn**      |        |        |        | ... |       |

O array pode ser transformado em um dataframe para facilitar a manipulação dos dados. Sendo assim, o array de componentes da PCA foi transformada em um dataframe e os dados utilizados para plotar o gráfico de loadings.
