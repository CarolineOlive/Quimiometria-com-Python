A one-way ANOVA uses the following null and alternative hypotheses:

H0 (null hypothesis): 1 = 2 = 3 =  = k (all the population means are equal)

H1 (null hypothesis): at least one population mean is different from the rest

Since the p-value is not less than 0.05, we fail to reject the null hypothesis. This means we do not have sufficient evidence to say that there is a difference in exam scores among the groups

```python
#importando os dados
dados_FQ = pd.read_excel('/content/drive/MyDrive/dados kaggle/matrizesnumricas/MatrizCaju FQ.xls')
dados_FQ.head()
```

O dados representam anlises de pH, fenis totais, acidez e grau brix realizadas em triplicata a partir de 10 amostras de caju.

[**Pr-processamento**](https://scikit-learn.org/stable/modules/preprocessing.html#)

A padronizao de conjuntos de dados  um requisito comum para muitos estimadores de aprendizado de mquina implementados no scikit-learn; eles podem se comportar mal se os recursos individuais no se parecem mais ou menos com dados normalmente distribudos padro: Gaussiano com mdia zero e varincia unitria.

Na prtica, muitas vezes ignoramos a forma da distribuio e apenas transformamos os dados para centraliz-los removendo o valor mdio de cada recurso e, em seguida, dimensionamos dividindo os recursos no constantes por seu desvio padro.

Por exemplo, muitos elementos usados na funo objetivo de um algoritmo de aprendizagem (como o kernel RBF de Support Vector Machines ou os regularizadores l1 e l2 de modelos lineares) assumem que todos os recursos esto centrados em torno de zero e tm varincia na mesma ordem. Se uma caracterstica tem uma varincia que  ordens de magnitude maior do que outras, ela pode dominar a funo objetivo e tornar o estimador incapaz de aprender com outras caractersticas corretamente como esperado.

O mdulo de pr-processamento fornece a classe de utilitrio [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler), que  uma maneira rpida e fcil de realizar a seguinte operao em um conjunto de dados do tipo array:





```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
#rodando o modelo de PCA sem pr-processamento dos dados
X = dados_FQ.loc[:, ['pH', 'Fenis Totais', 'Acidez', 'Grau Brix']]
pca = PCA(n_components=2)
pca.fit(X)
PCA(n_components=2)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
#pr-processamento
from sklearn.preprocessing import StandardScaler

#criando uma lista com os nomes das variveis
features = ['pH', 'Fenis Totais', 'Acidez', 'Grau Brix']

#separando as variveis
x = dados_FQ.loc[:, features].values

#separando os rtulos das amostras
y = dados_FQ.loc[:, ['Amostras de Caju']].values

#padronizando as variveis
x=StandardScaler().fit_transform(x)
```

```python
#gerando a PCA com o numero de componentes definida

pca = PCA(n_components=4) #definindo nmero de componentes

principalComponents = pca.fit_transform(x) #transformando os dados

principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2', 'PC3', 'PC4']) #criando um novo df com dimenso reduzida
```

```python
#concatenando os dados com os rtulos
finalDf = pd.concat([principalDf, dados_FQ[['Amostras de Caju']]], axis=1) 
finalDf.head()
```

```python
#eliminando a informao da replicata dos rotolos das amostras
finalDf['Amostras de Caju'] = finalDf['Amostras de Caju'].str[:3].str.replace(".", "")
```

```python
#obtendo a varincia explicada
PC1, PC2, PC3, PC4 = pca.explained_variance_ratio_
print(f'A varincia explicada de cada componente principal : \nPC1: {PC1*100}%\nPC2: {PC2*100}%\nPC3: {PC3*100}%\nPC4: {PC4*100}%')
```

```python
#plotando grfico de scores
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC1 (60.97%)', fontsize = 15)
ax.set_ylabel('PC2 (23.77%)', fontsize = 15)
ax.set_title('Grfico de Scores de PCA', fontsize = 20)

targets = ['A6', 'B9', 'C76', 'C10', 'B89', 'B65', 'B26', 'B74', 'B75', 'E50']
colors = ['lightpink', 'lightsteelblue', 'lightslategrey', 
          'lightblue', 'lightcyan', 'lightseagreen', 
          'lightgreen', 'darkgoldenrod', 'darkorange', 
          'darkred']

for Amostras, color in zip(targets, colors):
    indicesToKeep = finalDf['Amostras de Caju'] == Amostras
    ax.scatter(finalDf.loc[indicesToKeep, 'PC1'], finalDf.loc[indicesToKeep, 'PC2'], c = color, s=50)
ax.legend(targets)
ax.grid()
```

```python
#visualizando a tabelas de componentes e variveis que formam os loadings
componentes = pca.components_
componentes
```

```python
#criando um novo dataframe de loadings
load = pd.DataFrame(data=componentes, columns=features, index=['PC1', 'PC2', 'PC3', 'PC4'])
load
```

Um array  um arranjo de dados que representa linhas e colunas.
Cada lista (representada pelos colchetes) representa uma linha e cada linha corresponde a uma componente principal. As colunas represetam cada uma das variveis, como representado abaixo.

`Array(componentes, variveis)`



|  | **variavel1** | **variavel2** | **variavel3** | ... | **variaveln** |
|:------------:|:----------:|:-----------:|:-----------:|-----|:-------:|
|       **PC1**      |        |        |        | ... |       |
|       **PC2**      |        |        |        | ... |       |
|       **PC3**      |        |        |        | ... |       |
|       **...**      |        |        |        | ... |       |
|       **PCn**      |        |        |        | ... |       |

O array pode ser transformado em um dataframe para facilitar a manipulao dos dados. Sendo assim, o array de componentes da PCA foi transformada em um dataframe e os dados utilizados para plotar o grfico de loadings.

```python
#transpondo a matriz de loadings para plotar o grfico
load = load.T
load
```

```python
#plotando os dados de loadings
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC1 (60.97%)', fontsize = 15)
ax.set_ylabel('PC2 (23.77%)', fontsize = 15)
ax.set_title('Grfico de loadings de PCA', fontsize = 20)

variaveis = ['pH', 'Fenis Totais', 'Acidez', 'Grau Brix']
color = ['lightgreen', 'darkgoldenrod', 'darkorange', 
          'darkred']

ax = sns.scatterplot(data=load, x='PC1', y='PC2', hue=load.index, palette=color)
ax.grid()
```
